```python
hours_worked = 50
level = 1
wage = 100
salary = 0

#salary calculation
if hours_worked > 40:
    salary = 40 * wage + (hours_worked - 40) * wage * 1.5
    level += 1
else:
    salary = hours_worked * wage
print(salary)
        
#bonus calculation
if level >= 5:
    salary += 0.5*salary
    salary += 200
elif level < 5 and level >= 2:
    salary += 0.2*salary
    salary += 10
else:
    salary = salary
print(salary)

#tax deduction
if salary > 5000:
    salary -= salary*0.3
else:
    salary -= salary*0.1

print(salary)
```
